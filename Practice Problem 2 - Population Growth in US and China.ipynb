### Problem: Compare population growth in the U.S. and China.
Compare them both in terms of absolute growth and relative growth.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("countries.csv")
data.head()
```

```python
china = data[data.country=='China']
china.head()
```

```python
us= data[data.country=="United States"]
us.head()
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.title("Population growth in China and Us")
plt.legend(["US","China"])
plt.ylabel("Population in millions")
plt.xlabel("Years")
plt.show()
```

```python
#comparison in relation to a specific point
us_growth = us.population/us.population.iloc[0]*100
china_growth=china.population/china.population.iloc[0]*100
```

```python
plt.plot(us.year,us_growth)
plt.plot(china.year,china_growth)
plt.title("Population growth in China and Us")
plt.legend(["US","China"])
plt.ylabel("PSopulation")
plt.xlabel("Years")
plt.show()
```

### Problem: Find the relationship between life expectancy and GDP (NOT GDP per capita).
Use plt.scatter() and np.log10().

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("countries.csv")
data.head()
```

```python
data["gdp"] = (data.gdpPerCapita*data.population)/10**9
```

```python
data_2007 = data[data.year==2007]
```

```python
plt.scatter(data_2007.gdp,data_2007.lifeExpectancy,5)
plt.title("Gdp vs life Expectancy")
plt.ylabel("Life Expectancy")
plt.xlabel("Gdp In Billions")
```

```python
plt.scatter(np.log10(data_2007.gdp),data_2007.lifeExpectancy,5)
plt.title("Gdp vs life Expectancy")
plt.ylabel("Life Expectancy")
plt.xlabel("Gdp")
```

```python
print("Correlation between Gdp and life expectancy : " + str(np.log10(data_2007.gdp).corr(data_2007.lifeExpectancy)))
print("Correlation between GdpPerCapita and Life expectancy: " + str(np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)))
```

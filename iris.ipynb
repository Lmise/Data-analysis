```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
data.info()
```

```python
data.Species.unique()
```

```python
Iris_setosa = data[data.Species=='Iris-setosa']
Iris_setosa.head()
```

```python
Iris_versicolor = data[data.Species=='Iris-versicolor']
Iris_versicolor.head()
```

```python
Iris_virginica = data[data.Species=='Iris-virginica']
Iris_virginica.head()
```

```python
plt.subplot(3,1,1)
plt.title("Distribution of SepalLengthCm")
plt.hist(Iris_setosa.SepalLengthCm,20,range=(0,20))
plt.ylabel("Iris_setosa")
plt.subplot(3,1,2)
plt.hist(Iris_versicolor.SepalLengthCm,20,range=(0,20))
plt.ylabel("Iris_versicolor")
plt.subplot(3,1,3)
plt.hist(Iris_virginica.SepalLengthCm,20,range=(0,20))
plt.ylabel("Iris_virginica")
plt.show()
```

```python
plt.subplot(3,1,1)
plt.title("Distribution of SepalWidthCm")
plt.hist(Iris_setosa.SepalWidthCm,20,range=(0,20))
plt.ylabel("Iris_setosa")
plt.subplot(3,1,2)
plt.hist(Iris_versicolor.SepalWidthCm,20,range=(0,20))
plt.ylabel("Iris_versicolor")
plt.subplot(3,1,3)
plt.hist(Iris_virginica.SepalWidthCm,20,range=(0,20))
plt.ylabel("Iris_virginica")
plt.show()
```

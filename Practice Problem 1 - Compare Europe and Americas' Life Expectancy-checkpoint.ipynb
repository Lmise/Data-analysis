```python
#libraaries to be used
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Loading data from a csv using pandas
data = pd.read_csv('countries.csv')
data.head()
```

```python
#More info including data types, missing values,column names e.t.c
data.info()
```

```python
#1997 data
data_1997 = data[data.year==1997]
data_1997.head()
```

```python
Americas= data_1997[data_1997.continent== 'Americas']
Americas.head()
```

```python
Europe = data_1997[data_1997.continent=='Europe']
Europe.head()
```

```python
plt.subplot(2,1,1)
plt.title("Distribution of life expectancy")
plt.hist(Europe.lifeExpectancy,20,range=(50,100))
plt.ylabel("Europe")
plt.subplot(2,1,2)
plt.hist(Americas.lifeExpectancy,20,range=(50,100))
plt.ylabel("Americas")
plt.show()
```
